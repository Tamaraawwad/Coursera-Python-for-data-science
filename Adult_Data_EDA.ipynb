```python
#Write a Python program to display first 5 rows from Adult dataset.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("adult.csv")
df.info()
df.head()
```

```python
#How many men and women (sex feature) are represented in this dataset?
df['gender'].value_counts()
```

```python
#What is the average age (age feature) of women?
groubby_gender=df.groupby('gender')
groubby_gender['age'].mean()
```

```python
#What is the percentage of German citizens (native-country feature)?
print(df['native-country'].unique())
```

```python
# Remove leading and trailing spaces from 'native-country' column
df['native-country'] = df['native-country'].str.strip()

# Exclude rows where 'native-country' is '?' (indicating unknown)
df_cleaned = df[df['native-country'] != '?']

# Count occurrences of 'Germany'
german_count = df_cleaned['native-country'].value_counts().get('Germany', 0)

# Calculate percentage
german_percentage = (german_count / len(df_cleaned)) * 100

print(f"Percentage of German citizens: {german_percentage:.2f}%")
```

```python
#What are the mean and standard deviation of age for those who earn more than 50K per year (salary feature) and those who earn less than 50K per year?
df.groupby('income')['age'].agg(['mean','std'])
```

```python
#Is it true that people who earn more than 50K have at least a high school education?
 #(education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters or Doctorate feature)
#df.groupby('income')['education'].value_counts()
#print(df['education'].unique())

# Define the education levels considered as "high school or better"
high_school_or_better = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']

# Filter the dataset for those earning more than 50K
high_income_df = df[df['income'] == '>50K']

# Check if anyone earning >50K does NOT have at least a high school education
low_education_count = high_income_df[~high_income_df['education'].isin(high_school_or_better)].shape[0]

# Print the result
if low_education_count == 0:
    print("True: All people earning more than 50K have at least a high school education.")
else:
    print(f"False: {low_education_count} people earning more than 50K do not have at least a high school education.")
```

```python
# Display age statistics for each race (race feature) and each gender (sex feature). Use groupby() and describe().
#First, Remove extra spaces from column values
df['race'] = df['race'].str.strip()
df['gender'] = df['gender'].str.strip()

age_stats = df.groupby(['race', 'gender'])['age'].describe()
print(age_stats)

#Find the maximum age of men of Amer-Indian-Eskimo race.
max_age = df[(df['race'] == 'Amer-Indian-Eskimo') & (df['gender'] == 'Male')]['age'].max()
print("The maximum age of Amer-Indian-Eskimo men is:", max_age)

```

```python
#Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (marital-status feature)?
# Consider as married those who have a marital-status starting with Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.

# Remove leading and trailing spaces
df['income'] = df['income'].str.strip()
df['marital-status'] = df['marital-status'].str.strip()
df['gender'] = df['gender'].str.strip()

# Define married statuses
married_statuses = {'Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse'}

# Classify marital status
df['marital_status_grouped'] = df['marital-status'].map(lambda x: 'Married' if x in married_statuses else 'Single')

# Filter only males
male_df = df[df['gender'] == 'Male']

# Calculate total and high earners per marital status
total_counts = male_df['marital_status_grouped'].value_counts()
high_earners = male_df[male_df['income'] == '>50K']['marital_status_grouped'].value_counts()

# Calculate proportion of high earners
proportion_high_earners = (high_earners / total_counts * 100).fillna(0)

# Display results
print("Total men in each group:\n", total_counts)
print("\nHigh earners in each group:\n", high_earners)
print("\nProportion of high earners (%):\n", proportion_high_earners)
```

```python
#What is the maximum number of hours a person works per week (hours-per-week feature)?
# step 1
max_hours = df['hours-per-week'].max()
print(f"Maximum hours worked per week: {max_hours}")

# Remove leading and trailing spaces
df['income'] = df['income'].str.strip()

#How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?
# Step 2: Filter the dataset for people who work that maximum number of hours
max_hours_df = df[df['hours-per-week'] == max_hours]

# Step 3: Count how many of them earn more than 50K
high_earners = max_hours_df[max_hours_df['income'] == '>50K'].shape[0]

# Step 4: Calculate the total number of people working maximum hours
total_count = max_hours_df.shape[0]

# Step 5: Calculate the percentage of high earners among them
percentage_high_earners = (high_earners / total_count) * 100

# Display the results
print(f"Total people who work {max_hours} hours per week: {total_count}")
print(f"Number of people who earn more than 50K: {high_earners}")
print(f"Percentage of high earners among them: {percentage_high_earners:.2f}%")

#print(df[df['hours-per-week'] == 99].head(10))
```

```python
#Count the average time of work (hours-per-week) for those who earn a little and a lot (salary) for each country (native-country).

# Clean up any extra spaces in 'income' and 'native-country' columns
df['income'] = df['income'].str.strip()
df['native-country'] = df['native-country'].str.strip()

# Exclude rows where 'native-country' is '?' (unknown)
df_cleaned = df[df['native-country'] != '?']

# Group the data by 'native-country' and 'income', and calculate the mean of 'hours-per-week'
average_hours_by_country = df_cleaned.groupby(['native-country', 'income'])['hours-per-week'].mean().reset_index()

# Print the result
print(average_hours_by_country)

# Filter for Japan's data
japan_avg_hours = average_hours_by_country[average_hours_by_country['native-country'] == 'Japan']
print(japan_avg_hours)
```

```python
#Find out the total number of hours worked and mean salary as per different occupations.

# Clean up any extra spaces in 'occupation' and 'income' columns
df['occupation'] = df['occupation'].str.strip()
df['income'] = df['income'].str.strip()

# Exclude rows where 'occupation' is '?' (indicating unknown)
df_cleaned = df[df['occupation'] != '?']

# Group the data by occupation
occupation_stats = df_cleaned.groupby('occupation').agg(
    total_hours_worked=('hours-per-week', 'sum'),  # Total hours worked per occupation
    mean_salary=('income', lambda x: (x == '>50K').mean())  # Mean salary, where 1 represents '>50K' and 0 represents '<=50K'
).reset_index()

# Print the result
print(occupation_stats)
```
